2. Equipment table

```python
%%sql
-- create equipment table
create table Equipment (
  EquipCode char(9) ,
  EquipDescripion varchar(20) ,
  Price float,
  primary key (EquipCode)
);
```

```python
%%sql
-- insert data into equipment table
insert into Equipment (EquipCode, EquipDescripion, Price) values
('E001', 'Nao robot', 4000),
('E005', 'PC', 300.00),
('E006', 'Pens', 1.50),
('E009', 'Baxter Robot', 53000.75);
```

3. Order table

```python
%%sql
-- create order table
create table OrderTable (
  OrderNo char(9) ,
  clientid char(9) ,
  OrderDate DATE ,
  primary key (OrderNo, ClientId)
  foreign key (ClientId) references Client(ClientId)
);
```

```python
%%sql
-- insert data into order table
insert into OrderTable (OrderNo, clientid, OrderDate) values
('O001', 'C022', '17-03-02'),
('O002', 'C001', '17-03-10');

```

```python
%%sql
-- show the order table
select * from OrderTable;
```

4. Equipment Purchased table

```python
%%sql
-- create equipment purchased table
create table EquipmentPurchased (
  EquipCode char(9) ,
  OrderNo INTEGER ,
  Quantity INTEGER ,
  primary key (EquipCode, OrderNo)
  foreign key (EquipCode) references Equipment(EquipCode)
  foreign key (OrderNo) references OrderTable(OrderNo)
);
```

```python
%%sql
-- insert data into equipment purchased table
insert into EquipmentPurchased (EquipCode, OrderNo, Quantity) values
('E001', 001, 20),
('E006', 001, 10),
('E006', 002, 5),
('E009', 002, 5);
```

```python
%%sql
-- show the equipment purchased table
select * from EquipmentPurchased;
```
